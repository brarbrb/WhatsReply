```python
from unsloth import FastLanguageModel
import torch
#text only
model, tokenizer = FastLanguageModel.from_pretrained(
    # model_name = "unsloth/llama-3.1-8b-bnb-4bit",  # pick a model you like
    model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_seq_length = 1024,
    load_in_4bit = False,          # requires bitsandbytes + GPU; set False on CPU
    dtype = torch.float16,        # float16 on GPU; use torch.float32 on CPU
    device_map = "auto",          # will place on GPU if available
)
```

```python
prompt = "You are a helpful assistant. Briefly explain what Unsloth does."
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.inference_mode():
    out = model.generate(**inputs, max_new_tokens=128)
print(tokenizer.decode(out[0], skip_special_tokens=True))
```
